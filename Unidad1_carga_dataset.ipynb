```python
# Prompt: cargar el dataset amazon_cells_labelled.txt ubicado en el directorio 'sentiment\ labelled\ sentences'
# arriba de todo comentar el prompt utilizado para generar el código.

import pandas as pd
import os

# Definir la ruta del archivo
file_path = 'sentiment labelled sentences/amazon_cells_labelled.txt'

# Verificar si existe y cargar el dataset
if os.path.exists(file_path):
    # El archivo no tiene encabezados y usa tabuladores como separador
    df_amazon = pd.read_csv(file_path, sep='\t', header=None, names=['text', 'score'])
    print("Dataset amazon_cells_labelled.txt cargado con éxito.")
else:
    print(f"Error: No se encontró el archivo en {file_path}")
```
